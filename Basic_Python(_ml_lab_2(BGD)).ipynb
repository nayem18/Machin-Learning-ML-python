```python
import numpy as np
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt
```

```python
x = np.array ([6, 8, 12, 14, 18])
y = np.array ([350, 775,1150, 1395, 1675])
```

```python
w = np.array ([1, 2])
h = []
error_list = []
epoch = 10
for j in range(epoch):
  # Error Calculation
  for i in range(x.shape[0]):
    result = x[i]*w[1] + w[0]
    h.append(result)


  # print(h)
  error = h - y
  squared_error = (h-y)**2
  avg_sq_error = np.average(squared_error)


  # print(h - y)
  # print((h - y)**2)
  # print(np.average((h-y)**2))

  # print(error)
  # print(squared_error)
  # print(avg_sq_error)
  error_list.append(avg_sq_error)

  # print(x.shape[0])
  # print(w)
  # print(w[0])
  # print(w[1])

  # Calculating gradient of w0

  alpha = 0.001
  # dw0_sum = 0
  # for i in range(x.shape[0]):
  #   result = (h[i] - y[i])
  #   dw0_sum += result
  # print(dw0_sum)
  # dw0_sum = np.sum(h - y)
  # print(dw0_sum)
  # w[0] = w[0] -(alpha * 2 * dw0_sum) / x.shape[0]
  # print(w[0])

  dw1_sum = 0
  # for i in range(x.shape[0]):
  #   result = (h[i] - y[i])*x[i]
  #   dw1_sum += result
  # print(dw1_sum)

  dw0_sum = np.sum(h-y)
  w[0] = w[0] - (alpha * 2 * dw1_sum) / x.shape[0]

  dw1_sum = np.sum((h-y)*x)
  w[1] = w[1] - (alpha * 2 * dw1_sum) / x.shape[0]
  plt.plot(x, h)
  h.clear()
print(error_list)
  # print(a)
  # print(w)
print(w)
print(6*w[1] + w[0])
print(6 * 107 - 174.69)
# For drawing the graph
plt.scatter(x, y)  # scatter is (x, y ) dot

plt.xlabel("Pizza Size (Inch)")
plt.ylabel("Price (TK)")
plt.title("Linear Regression using BGD")
plt.grid()
plt.show()

plt.plot([i for i in range(epoch)], error_list)
plt.xlabel("Epoch")
plt.ylabel("Avg. sq. Error")
plt.title("Error Curve for BGD")
plt.grid()
plt.show()
```

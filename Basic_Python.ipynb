```python
import numpy as np
from matplotlib import pyplot as plt
# import  matplotlib.pyplot as plt
```

```python
bad = np.array ([[7, 7], [7, 4]])
good = np.array ([[3, 4], [1,4]])
test = np.array ([3, 7])
# print(bad)
# print(good)
# print(test)
```

```python
dist = []

# Distance from bad class data
for i in range(bad.shape[0]):
    distance = (bad[i] - test)**2
    distance =np.sqrt(distance[0] + distance[1])
    dist.append((distance, 'bad'))

# Distance from good class data
for i in range(good.shape[0]):
    distance = (good[i] - test)**2
    distance =np.sqrt(distance[0] + distance[1])
    dist.append((distance, 'good'))




print(dist)
dist.sort()
print(dist)

k = 3
count = 0
for i in range(k):
    if(dist[i][1] == 'good'):
        count =count  + 1
    else:
        count = count - 1
# print(count)
if (count > 0):
    print("Good Tissue")
else:
    print("Bad Tissue")

```

```python
center = (test[0], test[1])
radius = dist[k-1][0]

fig, ax = plt.subplots()
plt.scatter(good[0:2, 0], good[0:2, 1], color="blue", label="Good")
plt.scatter(bad[0:2, 0], bad[0:2, 1], color="red", label="Bad")
plt.scatter(test[0], test[1], color="green", label="Test")

circle = plt.Circle(center, radius, fill=False)
ax.add_artist(circle)

ax.set_xlim(-5, 15)
ax.set_ylim(-5, 15)
plt.legend()
plt.xlabel("Tissue Strength (kg/m^2)")
plt.ylabel("Acid Duration (sec)")
plt.title("Classification using KNN")

plt.grid()
PL

# print(good[0:2, 0])
# print(good[0:2, 1])
```

```python
import numpy as np  # alias
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt

x = np.array([6, 8, 12, 14, 18])
y = np.array([350, 775, 1150, 1395, 1675])
x_bar = np.average(x)
y_bar = np.average(y)
x_squared_bar = np.average(x**2)
xy_bar = np.dot(x, y)/x.shape[0]
m = (x_bar*y_bar - xy_bar)/(x_bar**2 - x_squared_bar)
c = y_bar - m*x_bar
print(m)
print(c)
# print(x.shape[0])
# print(y.shape[0])

```

```python
# for best fit line
dx = []
dy = []
for i in range(20):
  dx.append(i)
  dy.append(m*i + c)
```

```python
# plotting the graph
plt.scatter(x, y)
plt.plot(dx, dy, color="red")
plt.scatter([17], [m*17 + c], color="green")
plt.grid()
plt.show()
print(m*17+c)
```

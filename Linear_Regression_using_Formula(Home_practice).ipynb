```python
import numpy as np
from matplotlib import pyplot as plt

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,76,111,80,103,88,94,78,77,85,74])
```

```python
m = ((np.average(x)*np.average(y)) - (np.dot(x, y) / x.shape[0])) / ((np.average(x)**2 - np.average(x**2)))
c = np.average(y) - m* np.average(x)

print(m)
print(c)
```

```python
dx = []
dy = []
for i in range(20):
  dx.append(i)
  dy.append(m*i + c)
# END FOR
```

```python
plt.title('13 cars passing tollbooth.')
plt.scatter(x, y)
plt.scatter([15], [15*m + c], color = "orange")
plt.plot(dx, dy)
plt.xlabel('Size')
plt.ylabel('Price')
plt.grid()
plt.show()
```

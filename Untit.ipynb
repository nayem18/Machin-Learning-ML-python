```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([5, 9, 11,13, 15, 17])

m = ((np.average(x)*np.average(y)) - (np.dot(x, y) / x.shape[0])) / ((np.average(x)**2 - np.average(x**2)))
print(m)
c = (np.average(y) - m*np.average(x))
print(c)
```

```python
dx = []
dy = []
for i in range(20):
  dx.append(i)
  dy.append(m*i + c)
print(dx)
print(dy)
result = m*10 + c
print(result)
```

```python
plt.title("Leaniar rgrassion using formula.")
plt.scatter(x,y)
plt.scatter([10], [m*10 + c])
plt.xlabel("Size.")
plt.ylabel("Pizza price.")
plt.plot(dx, dy)
plt.grid()
```
